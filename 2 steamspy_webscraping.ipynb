```python
import requests
from bs4 import BeautifulSoup
```

```python
# find all data we need
url='https://steamspy.com/year/2021'
data=requests.get(url).content.decode('utf8')
soup=BeautifulSoup(data,'html.parser')
content=soup.find('tbody')
```

```python
# extract all data and put them into respective lists 
content1=content.find_all('a')
titlelist=[]
for i in range(len(content1)): 
    title=content1[i].text 
    titlelist.append(title)    

content2=content.find_all('td',{'class':'treleasedate'})
datelist=[]
for i in range(len(content2)): 
    date=content2[i].text 
    datelist.append(date)

content3=content.find_all('td',{'class':'tprice'})
pricelist=[]
for i in range(len(content3)): 
    price=content3[i].text 
    pricelist.append(price)

trs = soup.find_all('tr')[1:]
ownerlist=[]
for tr in trs:
    tds=tr.find_all('td')
    #print(len(tds))
    owner=tds[5].text
    ownerlist.append(owner)    

content5=content.find_all('td',{'class':'tplaytime'})
playtimelist=[]
for i in range(len(content5)): 
    playtime=content5[i].text 
    playtimelist.append(playtime)
    
trs = soup.find_all('tr')[1:]
publisherlist=[]
for tr in trs:
    tds=tr.find_all('td')
    #print(len(tds))
    publisher=tds[8].text
    publisherlist.append(publisher)   
```

```python
# put all the lists in a dictionary and export it in CSV format
import pandas as pd
dict1={'Title':titlelist, 'Date':datelist, 'Price':pricelist,'Owners':ownerlist, 'Playtime':playtimelist, 'Publisher':publisherlist}
df=pd.DataFrame(dict1)
df.to_csv('STEAMSPY2021.csv')
```
